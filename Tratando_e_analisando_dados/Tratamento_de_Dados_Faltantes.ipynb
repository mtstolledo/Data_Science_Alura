## Tratamento de Dados Faltantes 

```python
import pandas as pd
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
dados
```

```python
# Se o dado  for nulo, ficará como False no Dataframe
dados.isnull()
```

```python
#contrário do método isnull()
dados.notnull()
```

```python
dados.info()
```

```python
dados['Valor'].isnull()
```

```python
dados[dados['Valor'].isnull()]
```

```python
A = dados.shape[0]
dados.dropna(subset = ['Valor'], inplace = True)
B = dados.shape[0]
A - B
```

```python
dados[dados['Valor'].isnull()]
```

```python
#Campos nulos na coluna "Condominio"
dados[dados['Condominio'].isnull()].shape[0]
```

```python
#Filtrando de acordo com o "Tipo" "Apartamento"
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
# "~" inverte a série booleana
A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
#fillna atribui valor aos campos nulos 
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
```

```python
dados.info() 
```

```python
dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```
